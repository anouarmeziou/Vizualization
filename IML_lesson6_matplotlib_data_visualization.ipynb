# Numeric data visualization

## Matplotlib library

```python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.pyplot.style.use('ggplot')

import pandas as pd
from sklearn import datasets
```

```python
%matplotlib inline
```

## Dataset loading

```python
iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.data
```

```python
iris.target_names
```

```python
iris_frame = pd.DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
```

```python
iris_frame.head()
```

## Diagramm

```python
type(iris_frame.target.value_counts())
```

```python
plt.pie(iris_frame.target.value_counts(), labels=iris['target_names'])
plt.show()
```

```python
plt.pie(iris_frame.target.value_counts(), 
        autopct='%1.1f%%', labels=iris['target_names'])
plt.show()
```

```python
plt.pie(iris_frame.target.value_counts(), 
        explode=[0, 0.2, 0], autopct='%1.1f%%',
        shadow=True, startangle=90, labels=iris['target_names'])
plt.show()
```

## Scatter plot

```python
plt.scatter(iris_frame.loc[iris_frame.target == 0,'sepal length (cm)'], 
           iris_frame.loc[iris_frame.target == 0,'sepal width (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('setosa')
```

```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['red', 'blue', 'orange'])
plt.scatter(iris_frame.loc[:,'sepal length (cm)'], 
            iris_frame.loc[:,'sepal width (cm)'], c = iris_frame.target, 
            cmap = colors)

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Iris data')
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1, 2, 1)
plt.scatter(iris_frame.loc[:,'sepal length (cm)'], 
            iris_frame.loc[:,'sepal width (cm)'], c = iris_frame.target, 
            cmap = colors)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.subplot(1, 2, 2)
plt.scatter(iris_frame.loc[:,'petal length (cm)'], 
            iris_frame.loc[:,'petal width (cm)'], c = iris_frame.target, 
            cmap = colors)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

## Histogramm

```python
iris_frame[iris_frame.target == 0].hist('sepal length (cm)')
```

```python
plt.hist(iris_frame.loc[iris_frame.target == 0,'sepal length (cm)'], bins=12)
plt.show()
```

```python
plt.figure(figsize=(16,4))

plt.subplot(1,3,1)
plt.hist(iris_frame.loc[iris_frame.target == 0,'sepal length (cm)'])
plt.title('setosa')

plt.subplot(132)
plt.hist(iris_frame.loc[iris_frame.target == 1,'sepal length (cm)'])
plt.title('versicolor')

plt.subplot(133)
plt.hist(iris_frame.loc[iris_frame.target == 2,'sepal length (cm)'])
plt.title('virginica')
```

```python
plt.title('Sepal length (cm)')
plt.hist(iris_frame.loc[iris_frame.target == 0,'sepal length (cm)'], 
         alpha=0.5, label='setosa')
plt.hist(iris_frame.loc[iris_frame.target == 1,'sepal length (cm)'], 
         alpha=0.5, label='versicolor')
plt.hist(iris_frame.loc[iris_frame.target == 2,'sepal length (cm)'], 
         alpha=0.5, label='virginica')
plt.legend(loc='upper right')

plt.savefig('nice_hist.png', dpi=500)
```

```python
plt.figure(figsize = (20, 24))

plot_number = 0
for feature_name in iris['feature_names']:
    for target in iris_frame.target.unique():
        plt.subplot(4, 3, plot_number + 1)
        plt.hist(iris_frame[iris_frame.target == target][feature_name])
        plt.title(iris['target_names'][target])
        plt.xlabel('cm')
        plt.ylabel(feature_name[:-4])
        plot_number += 1
```
