## Boston regression problem

Boston dataset visualization with help of Seaborn

### Data loading

```python
boston_set = datasets.load_boston()
boston_data = DataFrame(data=boston_set.data, columns=boston_set.feature_names)
boston_data['target'] = boston_set.target
```

```python
print(boston_set['DESCR'])
```

```python
boston_data.isna().sum()
```

### Target distrubution

```python
sns.distplot(boston_data.target)
```

```python
boston_data.describe()
```

### Correlations

Correlations helps us to figure out if there is some linear connections between the features and tagget

```python
boston_data.corr()
```

This idea is hard to perceive. Let's construct the heatmap table over the correlations.

```python
plt.figure(figsize=(15, 8))
sns.heatmap(boston_data.corr(), annot = True, cmap='RdBu_r')
```

Now you can clearly see that there is some features with pretty strong correlations with the target variable.
Let's analyse separately the $LSTAT$ feature, which has a significant correlation with the target variable.

### Plotting the best correlated feature

```python
sns.jointplot(data=boston_data, y='target', x='RM', 
              kind='reg', line_kws={'color':'r'})
```

The result corresponds to our expectations: there is a linear relationship  between the features and the target. 

### Plotting the worst correlated feature

Now let's analyze $DIS$ feature, which has the low correlation with target.
Let's make a residplot (residuals of linear regression) for this feature.

```python
sns.jointplot(data=boston_data, y='target', x='DIS', 
              kind='reg', line_kws={'color':'r'})
```

```python
sns.residplot(data=boston_data, y='target', x='DIS')
```

You can see that the residuals of $ DIS $ feature have some structure, which is the reason to conduct additional experiments with this feature. For example, let's take a look at lmplot (data with a regression model).

We have already tried the linear model, now let's try order = 2 and order = 3

```python
sns.lmplot(data=boston_data, x='DIS', y='target', order=2,
           line_kws={'color':'r'})

sns.lmplot(data=boston_data, x='DIS', y='target', order=3,
           line_kws={'color':'r'})
```

### Pairplot for the observations

Let's construct pairplot over a part of features (for all data the picture will not be convenient for analysis)

```python
sns.pairplot(boston_data.iloc[:, 4:9])
```

Now you can see that there are a lot of significant non-linear connections between the features, and there are some outliers in terms of the $NOX$ feature (outstanding points on all scatter plots with $NOX$).
